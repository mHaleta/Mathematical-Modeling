```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

# Побудова математичних моделей експериментально-статистичними методами

```python
X = np.arange(1, 9)
Y = np.array([3.6, 3.63, 3.53, 3.63, 3.58, 3.65, 3.7, 3.6])
```

```python
plt.figure()
plt.scatter(X,Y)
plt.grid(True)
plt.show()
```

## Лінійна регресія $\ \ \mathbf{y=kx+b}$

```python
Z = np.polyfit(X, Y, deg=1)
f = np.poly1d(Z)
Y1 = f(X)
```

```python
plt.figure()
plt.scatter(X, Y)
plt.plot(X, Y1)
plt.grid(True)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
F = np.sum((Y1 - np.mean(Y1))**2)*(X.shape[0]-2)/np.sum((Y-Y1)**2)
F_table = 2.365
if F > F_table:
    print("Модель адекватна")
else:
    print("Модель неадеватна")
```

## Квадратична регресія $\ \ \mathbf{y=a_0x^2+a_1x+a_2}$

```python
Z = np.polyfit(X, Y, deg=2)
f = np.poly1d(Z)
Y2 = f(X)
```

```python
plt.figure()
plt.scatter(X, Y)
plt.plot(X, Y2)
plt.grid(True)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
F = np.sum((Y2 - np.mean(Y2))**2)*(X.shape[0]-2)/np.sum((Y-Y2)**2)
F_table = 2.365
if F > F_table:
    print("Модель адекватна")
else:
    print("Модель неадеватна")
```

## Показникова регресія $\ \ \mathbf{y=a_0e^{a_1x}}$

```python
A = np.array([[X.shape[0], np.sum(X)], [np.sum(X), np.sum(X**2)]])
b = np.array([np.sum(np.log(Y)), np.sum(np.log(Y)*X)])

coeffs = np.linalg.solve(A, b)

a_0 = np.exp(coeffs[0])
a_1 = coeffs[1]

Y3 = a_0*np.exp(a_1*X)
```

```python
plt.figure()
plt.scatter(X, Y)
plt.plot(X, Y3)
plt.grid(True)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
F = np.sum((Y3 - np.mean(Y3))**2)*(X.shape[0]-2)/np.sum((Y-Y3)**2)
F_table = 2.365
if F > F_table:
    print("Модель адекватна")
else:
    print("Модель неадеватна")
```

## Гіперболічна регресія $\ \ \mathbf{y=a_0+\frac{a_1}{x}}$

```python
A = np.array([[X.shape[0], np.sum(1/X)], [np.sum(1/X), np.sum(1/(X**2))]])
b = np.array([np.sum(Y), np.sum(Y/X)])

coeffs = np.linalg.solve(A, b)

a_0 = coeffs[0]
a_1 = coeffs[1]

Y4 = a_0 + a_1/X
```

```python
plt.figure()
plt.scatter(X, Y)
plt.plot(X, Y4)
plt.grid(True)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
F = np.sum((Y4 - np.mean(Y4))**2)*(X.shape[0]-2)/np.sum((Y-Y4)**2)
F_table = 2.365
if F > F_table:
    print("Модель адекватна")
else:
    print("Модель неадеватна")
```
